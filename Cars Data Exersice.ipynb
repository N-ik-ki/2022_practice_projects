**9.Print count of total cars per company**

```python
#Print count of total cars per company
cars=df.groupby('company').company.count()
print(cars)
#df['company'].value_counts()
```

--> The “group by” process: split-apply-combine

(1) Splitting the data into groups. (2). Applying a function(count,sum,mx,mn) to each group independently, (3) Combining the results into a data structure.
